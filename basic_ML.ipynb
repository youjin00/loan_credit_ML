```python
import pandas as pd
import numpy as np

df['Mean_Credit'] = np.mean(df[['HighCredit', 'LowCredit']], axis=1)
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()

df['대출등급'] = label_encoder.fit_transform(df['대출등급'])
df.head()
```

```python
df_label = df.copy()
df_onehot = df.copy()
ntrain = 96294
train = df[:ntrain]

y_train = train[['대출등급']].copy()
```

```python
drop_features = ['ID', '대출등급']
df_label = df_label.drop(drop_features, axis=1).copy()
df_label.head()
```

```python
drop_features = ['ID', '대출등급']
df_onehot = df_onehot.drop(drop_features, axis=1).copy()
df_onehot.head()
```

```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder

df_label = df_label.apply(LabelEncoder().fit_transform)
df_label.head() #float형 데이터 변환
```

```python
onehot_cols = df_label.columns.tolist()
df_onehot = pd.get_dummies(df_label, columns=onehot_cols)

df_onehot.head()
```

```python
import pickle
with open('credit_feature_encoding.pickle', 'wb') as pickle_filename :
    pickle.dump(df_onehot, pickle_filename)

import pickle
with open('credit_feature_encoding_y.pickle', 'wb') as pickle_filename :
    pickle.dump(y_train, pickle_filename)
```

```python
import warnings
warnings.filterwarnings('ignore')

import pickle
with open('credit_feature_encoding.pickle', 'rb') as pickle_filename :
    df_onehot = pickle.load(pickle_filename)

import pickle
with open('credit_feature_encoding_y.pickle', 'rb') as pickle_filename :
    y_train = pickle.load(pickle_filename)

ntrain = 96294
X_train, X_test = df_onehot[:ntrain], df_onehot[:ntrain]
X_train.head()
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score #학습 데이터 성능 우선 확인 가능

from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.naive_bayes import GaussianNB
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
```

```python
knn_model = KNeighborsClassifier()
logreg_model = LogisticRegression()
svc_model = SVC()
decision_model = DecisionTreeClassifier()
random_model = RandomForestClassifier()
extra_model = ExtraTreesClassifier()
gbm_model = GradientBoostingClassifier()
nb_model = GaussianNB()
xgb_model = XGBClassifier(eval_metirc='logloss') #워닝 메세지 방지용으로 넣어준 것
lgbm_model = LGBMClassifier()
#각각의 모델에 인자 값을 넣어줄 수 있음

models = [
    knn_model,
    logreg_model,
    svc_model,
    decision_model,
    random_model,
    extra_model,
    gbm_model,
    nb_model,
    xgb_model,
    lgbm_model
] #리스트에 넣어줌

k_fold = KFold(n_splits=10, shuffle=True, random_state=0) #KFold 사용
results = dict() #각각의 모델

for alg in models :
    alg.fit(X_train, y_train) #학습데이터와 정답 데이터를 넣어서 학습 시킨 것
    score = cross_val_score(alg, X_train, y_train.values.ravel(), cv=k_fold, scoring='accuracy') #cross_val을 활용해서 성능 확인
    results[alg.__class__.__name__] = np.mean(score)*100
```
